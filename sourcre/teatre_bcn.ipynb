```python
from bs4 import BeautifulSoup
import requests
from IPython.display import SVG, display

# Define the base URL for the initial page
#opcions: teatre, humor, musicals, infantil, dansa, experiències, proximitat, opera
#vull_veure = "humor"
#base_url = 'https://www.teatrebarcelona.com/genere/teatre' % (vull_veure)

base_url = url
# Create an empty list to store the scraped data
all_data = []

# Initialize a variable to track the current page number
current_page = 2

# Set the maximum number of pages to scrape
max_pages = 2  # Change this to the number of pages you want to scrape

while current_page <= max_pages:
    print("--------------------------------PAGE", current_page, "---------------------------------------")

    # Construct the URL for the current page
    page_url = f'{base_url}/pagina/{current_page}'

    # Send an HTTP GET request to the page
    response = requests.get(page_url)

    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract "funcion-item-list" elements
        funcion_items = soup.find_all('div', class_='funcion-item-list')

        # Loop through each "funcion-item-list" and extract the desired information
        for funcion_item in funcion_items: # for each item in the funcion item we can extratc info
            #print("\n")
            # TITLE
            #if image le'structura es fucnitom item -> class e -> class tool tip
            if funcion_item.find("div", class_="e"):

                title = funcion_item.find('a', {'class': 'tooltip'})['data-title']
            else:
                # If "e" does not exist, find the <div> element with class "ts" and extract the text
                title_div = funcion_item.find('div', class_='ts')
                title = title_div.get_text(strip=True)

            #print(title)

            # DATE
            tx_div = funcion_item.find('div', class_='tx')
            d_div = tx_div.find('div', class_='d')
            dates_span = d_div.find('span', class_='el-mobile').get_text()
            date_parts = dates_span.split("Fins:")  #borrem fins: i obtenim llista

            if len(date_parts) == 2 :
                des_de_date = date_parts[0].replace("Des de:", "").strip()
                fins_date = date_parts[1].strip()

            elif len(date_parts) == 1 :
                des_de_date = date_parts[0].replace("El", "").strip()
                fins_date = des_de_date

            #print("Des de: ", des_de_date)
            #print("Fins: ", fins_date)

            # PRICE (dicount / price / not available)
            price_dis = funcion_item.find('div', class_="p")
            price_div = funcion_item.find('div', class_="price")
            next_div = funcion_item.find('div', class_ ="next")


            if price_dis != None:
                actual_price = price_dis.find('span', class_="price").text
                discount = price_dis.find('span', class_="sale").text
                old_price = price_dis.find('div', class_="o").text.strip()
                price = actual_price
                #print("Preu: ", actual_price, discount, old_price)

            if price_div != None:
                price_span = price_div.find('span')
                price = price_span.get_text(strip=True)
                #print("Preu: ", price)

            if next_div != None:
                not_price = next_div.text.strip() # You can use .text to extract the text
                price = not_price
                #print("Preu: ", not_price)

        # RATE
            # Find the <div> element with class "palmas"
            palmas_div = funcion_item.find('div', class_='palmas')

            if palmas_div != None:

                # Find the <div> element with class "rate" within the "palmas" div
                rate_div = palmas_div.find('div', class_='rate')

                # Extract the rate percentage from the "rate" div
                rate_percentage = rate_div['style']

            if palmas_div == None:
                rate_percentage = "0%"
            #print("Rate Percentage:", rate_percentage)

        # LINK - more info per cada obra
            inf_link = funcion_item.find('a', class_='inf')
            href = inf_link['href']

            individual_response = requests.get(href)
            individual_soup = BeautifulSoup(individual_response.text, 'html.parser')

            # print(individual_soup.prettify())
            sinopsis_div = individual_soup.find('div', class_='synopsis el-desktop')
            presentation_div = individual_soup.find('div', class_='presentation')

            if sinopsis_div:
                p_elements = sinopsis_div.find_all('p')
                if len(p_elements) > 0:
                    sinopsis = ' '.join(p.get_text() for p in p_elements)
                else:
                    sinopsis = "Sinopsis not found"
            elif presentation_div:
                sinopsis = presentation_div.get_text()
            else:
                sinopsis = "Sinopsis not found"

            #print(sinopsis)

            # MES INFO - DURACIO / IDIOMA / EDAT

            info_div = individual_soup.find('div', class_='info')
            items = info_div.find_all('div', class_='item')

            for item in items:
                t_element = item.find('div', class_='t')
                if t_element:
                    text = t_element.get_text(strip=True)
                if "Durada:" in text:
                    durada = text.replace("Durada:", "").strip()
                elif "Idioma:" in text:
                    idioma = text.replace("Idioma:", "").strip()
                elif "Edat:" in text:
                    edat = text.replace("Edat:", "").strip()



            #print("Durada:", durada)
            #print("Idioma:", idioma)
            #print("Edat:", edat)

            # Store the extracted data in a dictionary
            data = {
                "Title": title,
                "Price": price,
                "Start Date": des_de_date,
                "End Date": fins_date,
                "rate": rate_percentage,
                "sinopsis": sinopsis,
                "durada" : durada,
                "idioma": idioma,
                "edat": edat
            }

            # Append the data to the all_data list
            all_data.append(data)

        # Increment the current_page variable to move to the next page
        current_page += 1
    else:
        #print(f"Failed to retrieve page {current_page}. Status code: {response.status_code}")
        break  # Exit the loop in case of an error

# Process the scraped data as needed
#for item in all_data:
    # Process the item
    #print(item)
result = pd.DataFrame(all_data)
print(result.head())
# Save the DataFrame to a CSV file
result.to_csv('your_file_humor.csv', index=False)
```
